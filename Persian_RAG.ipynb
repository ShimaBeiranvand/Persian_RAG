```python
def extract_text_from_pdf(path):

    normalizer = Normalizer()
    text_parts = []

    # extracting and normalizing text from pdf
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                normalized = normalizer.normalize(page_text)
                text_parts.append(normalized)
    text = "\n".join(text_parts)

    # removing the unreadable charachters which are witten in text as Q
    text = text.replace("Q", " ")

    # converting text from LTR to FTL
    new_text = ''
    for line in text.splitlines():
        # print('line:')
        # print(line)
        for char in line[::-1]:
            # print(char)
            new_text += char
            # print(new_text)
        new_text += ' '

    # # also storing the extracted text in a file for availability
    # output_dir = '/content/extracted_text'
    # os.makedirs(output_dir, exist_ok=True)  # create dir if it doesn't exist

    output_file_path = "/content/drive/MyDrive/extracted_text.txt"

    # Write extracted text to file
    with open(output_file_path, 'w', encoding='utf-8') as f:
        f.write(new_text)

    print(f"✅ Text extracted and saved to {output_file_path}")

    return text
```

```python
# chunking semanticly

from multilingual_sentence_tokenizer import sentence_tokenizer
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('sentence-transformers/LaBSE')

def semantic_chunker(text):

    # Tokenize into sentences
    sentences = sentence_tokenizer.tokenize(text, language='fa')

    # # Print the sentences
    # for sentence in sentences:
    #     print(sentence)
    #     print('***************************************************************')

    # embedding the sentences
    embeddings = model.encode(sentences)
    # Print the embeddings
    # for embedding in embeddings:
    #     print(embedding)
    #     print('***************************************************************')

    # measuring cosine similarity and chunking based on similar sentences
    similarity_threshold = 0
    max_chunk_size = 3

    chunks = []
    current_chunk = [sentences[0]]

    for i in range(1, len(sentences)):
        sim = cosine_similarity(
            [embeddings[i]],
            [embeddings[sentences.index(current_chunk[-1])]]
        )[0][0]

        if sim >= similarity_threshold and len(current_chunk) < max_chunk_size:
            current_chunk.append(sentences[i])
        else:
            chunks.append(' '.join(current_chunk))
            current_chunk = [sentences[i]]

    # Append the last chunk
    if current_chunk:
        chunks.append(' '.join(current_chunk))

    # print chunks
    for idx, chunk in enumerate(chunks, 1):
        print(f"Chunk {idx}:\n{chunk}\n")

    return chunks
```

```python
# second retriever code

from rank_bm25 import BM25Okapi
from sklearn.preprocessing import minmax_scale

class HybridRetriever:
    def __init__(self, chunks, embedding_model):
        self.chunks = chunks
        self.embedding_model = embedding_model

        # Prepare FAISS index
        self.embeddings = self.embed_chunks(chunks)
        self.faiss_index = self.build_faiss_index(self.embeddings)

        # Prepare BM25 index
        self.tokenized_chunks = [chunk.split() for chunk in chunks]
        self.bm25 = BM25Okapi(self.tokenized_chunks)

    def embed_chunks(self, chunks):
        return self.embedding_model.encode(chunks, convert_to_tensor=False, show_progress_bar=True)

    def build_faiss_index(self, embeddings):
        dim = len(embeddings[0])
        index = faiss.IndexFlatIP(dim)
        faiss.normalize_L2(embeddings)
        index.add(embeddings)
        return index

    def rank_faiss(self, query, top_k=5):
        query_vec = self.embedding_model.encode([query])
        faiss.normalize_L2(query_vec)
        D, I = self.faiss_index.search(query_vec, len(self.chunks))
        return list(zip(I[0], D[0]))  # (index, score)

    def rank_bm25(self, query):
        tokenized_query = query.split()
        scores = self.bm25.get_scores(tokenized_query)
        return list(enumerate(scores))  # (index, score)

    def fuse_rankings(self, faiss_ranks, bm25_ranks, top_k=5):
        # Convert to dict for easy lookup
        faiss_dict = dict(faiss_ranks)
        bm25_dict = dict(bm25_ranks)

        # Normalize scores
        faiss_scores = minmax_scale(list(faiss_dict.values()))
        bm25_scores = minmax_scale(list(bm25_dict.values()))

        fused_scores = {}
        for idx in range(len(self.chunks)):
            fused_score = faiss_scores[idx] + bm25_scores[idx]  # simple sum fusion
            fused_scores[idx] = fused_score

        # Sort by fused score descending
        sorted_fused = sorted(fused_scores.items(), key=lambda x: x[1], reverse=True)

        # Return top K chunks
        return [(self.chunks[idx], score) for idx, score in sorted_fused[:top_k]]

    def retrieve(self, query, top_k=3):
        faiss_ranks = self.rank_faiss(query)
        bm25_ranks = self.rank_bm25(query)
        top_results = self.fuse_rankings(faiss_ranks, bm25_ranks, top_k)
        return top_results

```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, GenerationConfig, BitsAndBytesConfig

# Model name on Hugging Face
model_name = "MehdiHosseiniMoghadam/AVA-Llama-3-V2"

# Load tokenizer
llama_tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define quantization config for 8-bit loading
quantization_config = BitsAndBytesConfig(
    load_in_8bit=True
)

# Load model with quantization and device map
llama_model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    low_cpu_mem_usage=True,
    quantization_config=quantization_config
)

```

```python
# QA generation function for LLaMA model
def ask_llama(question: str, context: str = "") -> str:
    prompt = f"""شما یک دستیار هوش مصنوعی مفید هستید. از متن زیر برای پاسخ به سؤال استفاده کن.
اگر پاسخ در متن وجود نداشت، بگو "نمی‌دانم."

### متن:
{context}

### سؤال:
{question}

### پاسخ:"""

    inputs = llama_tokenizer(prompt, return_tensors="pt").to(llama_model.device)

    gen_config = GenerationConfig(
        do_sample=False,  # deterministic — for QA, sampling can cause hallucination
        max_new_tokens=150,  # longer in case of more elaborate answers
        pad_token_id=llama_tokenizer.eos_token_id
    )

    output = llama_model.generate(**inputs, generation_config=gen_config)
    answer = llama_tokenizer.decode(output[0], skip_special_tokens=True)

    # Optionally, extract the answer only after '### Answer:' if your model's decoding includes prompt text
    answer_part = answer.split("### Answer:")[-1].strip()

    return answer_part

```

```python
# fixing the order of numbers generated by model because it shows them wrong

import re

def fix_persian_numbers_in_text(text):
    persian_digits = '۰۱۲۳۴۵۶۷۸۹'

    def reverse_persian_num(match):
        num_str = match.group()
        reversed_num = ''.join(num_str[::-1])
        return reversed_num

    # Regex to find sequences of Persian digits
    pattern = f'[{persian_digits}]+'
    fixed_text = re.sub(pattern, reverse_persian_num, text)
    return fixed_text

# Examples:
print(fix_persian_numbers_in_text("عمق آب ۰۳ متر است"))  # Output: "عمق آب ۳۰ متر است"
print(fix_persian_numbers_in_text("شماره ۱۲۳۴"))        # Output: "شماره ۴۳۲۱"

```

```python
# all in one function

def persian_rag(question , semantic_chunks):

    # retriever
    top_docs = retriever.retrieve(question, top_k=3)

    context = " ".join([chunk for chunk, score in top_docs])

    # answer generation with llama
    llama_answer = ask_llama(question=question, context=context)

    return llama_answer , top_docs
```

```python
import gradio as gr

def qa_system(question):
    answer, top_docs = persian_rag(question, semantic_chunks)
    docs_text = "\n".join([f"- {chunk}" for chunk, score in top_docs])
    lines = answer.splitlines()
    for line in lines:
        if line.startswith("### پاسخ:"):
            short_answer = line.replace("### پاسخ:", "").strip()
    return f"Answer:\n{short_answer}\n\nTop Documents:\n{docs_text}"

gr.Interface(fn=qa_system, inputs="text", outputs="text").launch()
```
